```python
import pandas as pd
import numpy as np
```

```python
tarifas = pd.read_csv('TARIFAS_ACUEDUCTO_Y_ALCANTARILLADO_2021_20250523.csv')
tarifas
#Linea de codigo para visualizaci√≥n de los datos
```

```python
promediodecargofijo=tarifas['CARGO FIJO'].mean()
promediodecargofijo
#Esta linea de codigo nos dice cual es el promedio de la columna de cargos fijos, segun servicios; acueducto y alcantarillado
```

```python
filtro1 = tarifas[tarifas['SECTOR RESIDENCIAL'] == 'COMERCIAL']
filtro1
#Esta linea de codigo nos permite filtrar los sectores residenciales seleccionados como comerciales
```

```python
tarifas['sen'] = np.sin(tarifas['CARGO FIJO'])
tarifas['sen']
```

```python
tarifas['cos'] = np.cos(tarifas['CARGO FIJO'])
tarifas['cos']
```

```python
tarifas['raiz'] = np.sqrt(tarifas['CARGO FIJO'])
tarifas['raiz']
```

```python
tarifas['sumaacumuladacargofijo'] =tarifas['CARGO FIJO'].cumsum()
tarifas['sumaacumuladacargofijo']
```

```python
tarifas['DESVIACION'] =tarifas['CARGO FIJO'].std()
tarifas['DESVIACION']
```

```python
tarifas['varianza'] =tarifas['CARGO FIJO'].var()
tarifas['varianza']
```

```python
tarifas['minimo'] =tarifas['CARGO FIJO'].min()
tarifas['minimo']
```

```python
tarifas['maximo'] =tarifas['CARGO FIJO'].max()
tarifas['maximo']
```
